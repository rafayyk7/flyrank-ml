## 3. Success metric

*One metric you can defend. What number means 'good'?*

Our primary, highly-defensible business success metric is **Precision@K** (specifically, **Precision@50**).

* **Why this metric?** Since editorial capacity is limited to about 50 page rewrites a week, standard classification metrics like Recall or ROC-AUC are less relevant to our stakeholders. We only care about the quality of the top 50 recommendations we hand to writers.
* **What number means "good"?** A Precision@50 of **70% or higher** is our target. This means at least 35 out of the 50 pages we assign to writers are genuinely decaying and worth updating, dramatically reducing wasted creative writing budget.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple, hard-coded rule like `if age > 180 days and current_traffic < peak_traffic * 0.8: refresh` fails in the real world for several reasons:

1. **Varying Lifespans:** Highly trendy "viral" pages decay within days, while cornerstone "evergreen" pages can remain perfectly healthy for years without any updates.
2. **Seasonality Noise:** A sudden 20% traffic drop during winter holidays isn't actual content decay; a fixed rule would falsely flag it as broken, wasting valuable writing budget.
3. **Complex Feature Interactions:** Real decay is a subtle interaction of multiple parameters (e.g., CTR decline, impression share drops, position shifts, and historical client volatility). Machine learning can weight these non-linear relationships dynamically to make robust predictions that hard-coded `if/else` rules simply cannot replicate.
