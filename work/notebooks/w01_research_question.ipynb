## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

By loading and analyzing the starter dataset, we can ground our decision framework in real-world scale numbers.

We load the bundled dataset to compute:
1. The total size of our baseline page catalog.
2. The count of pages that have historically dropped in visibility and been labeled as "declining".
3. The overall baseline decay rate of the tracked search footprint.

```python
import os
import sys
import subprocess
import pandas as pd

# 1. If we are running in Google Colab, clone your repo to get the data files
if 'google.colab' in sys.modules:
    REPO_URL = "https://github.com/rafayyk7/flyrank-ml.git"
    REPO_DIR = "flyrank-ml"

    if not os.path.exists(REPO_DIR):
        print("Cloning repository into Google Colab...")
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

    # Move inside the cloned repository directory where the data folder lives
    os.chdir(REPO_DIR)
    print(f"Current working directory updated to: {os.getcwd()}")

# 2. Try the two common paths for the CSV
csv_path = 'data/raw/content_refresh_anonymized.csv'
if not os.path.exists(csv_path):
    csv_path = '../../data/raw/content_refresh_anonymized.csv'

# 3. Load the dataset
print("Loading data...")
df = pd.read_csv(csv_path)

# 4. Compute the exact numbers to back your lane
total_pages = len(df)
is_declining = df['trend_direction'].str.lower().eq('down').astype(int)
decayed_pages = is_declining.sum()
decay_rate = (decayed_pages / total_pages) * 100

print("\n--- RESULTS ---")
print(f"Number 1: Total pages analyzed in starter set = {total_pages:,}")
print(f"Number 2: Pages flagged as decaying (trend_direction = 'down') = {decayed_pages:,}")
print(f"Number 3: Base baseline decay rate of this search ecosystem = {decay_rate:.2f}%")
```
